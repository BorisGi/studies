**3. Основное таргет значение: 1 - сделка была обманной, 0 - честной. Сделать график barplot для этих двух классов из колонки 'Class'. Посчитать количество элементов в классе 1 и в классе 2.**

```python
print('Количество честных сделок:', dataset[dataset.Class == 0].Class.count())
print('Количество обманных сделок:', dataset[dataset.Class == 1].Class.count())
```

**4. Сколько денег было использовано в обманных операциях, а сколько - в нормальных операциях?** Сделать частотный график по колонке Amount для колонки Class по двум независимым выборкам: обманная операция, нормальная операция

```python
honest = dataset[dataset.Class == 0]
dishonest = dataset[dataset.Class == 1]
```

```python
print('Общая сумма всех честных сделок:', round(honest.Amount.sum(), 2))
print('Общая сумма всех обманных сделок:', round(dishonest.Amount.sum(), 2))
```

```python
plt.rcParams['figure.figsize']=(14,4)
f, (ax1, ax2) = plt.subplots(1,2)


sns.distplot(honest.Amount, bins=50, kde=False,
             hist_kws={'color':'#052A6E', 'edgecolor':'k', 'linewidth':'1'},             
                       ax=ax1
                       )

sns.distplot(dishonest.Amount, bins=50, kde=False,
             hist_kws={'color':'#FFA500', 'edgecolor':'k', 'linewidth':'1'},             
             ax=ax2
            )
ax1.set_title('Честные сделки', fontsize=16)
ax2.set_title('Мошеннические', fontsize=16)
plt.show()
```

**Сделать тест Стьюдента для двух выборок:**

```python
stats.ttest_ind(honest['Amount'], dishonest['Amount'])
```

Выборки не похожи, P-value значительно меньше 0.05.

**6. Построить матрицу корреляций для всех значений**

```python
f, ax = plt.subplots(figsize=(20, 20))
corr = np.round_(dataset.corr(), decimals=2)
sns.heatmap(corr, cmap='coolwarm', linewidths=0.2, annot=True)
```

Аттрибутов, коррелирующих аттрибутов больше чем на 95%, нет.

```python
# создаем матрицу корреляций
corr_matrix = dataset.corr().abs()

# Выбираем верхний треугольник матрицы
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Ищем индексы колонок с корреляцией большей чем 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

print(to_drop)
```

**7. Взять все имеющиеся атрибуты из таблицы для предсказания. Разделить на тренировочные и тестовые. Написать модель деревья решений. Сделать classification report.**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

from sklearn import preprocessing
```

Модель Random Forest

```python
output_y = dataset.iloc[:,29:]
input_x = dataset.iloc[:,1:29]
```

```python
le=LabelEncoder()
targets=(output_y.iloc[:,:]==1).idxmax(1)
print(targets.value_counts())
Y=le.fit_transform(targets)
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_x, Y, test_size=0.3)
```

```python
# создаем модель деревья решений
# выбираем 100 деревьев в качестве параметра
model=RandomForestClassifier(n_estimators=100)

# обучаем модель
model.fit(X_train, y_train)
```

```python
headers = list(X_train.columns.values)

feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(6, 6))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Важность атрибутов')
plt.ylabel('Атрибуты')
plt.title("Наиболее важные атрибуты")
plt.legend()
plt.show()
```

Как делать classification report:

```python
model_pred = model.predict(X_test)

print(classification_report(y_test, model_pred))

model.score(X_test, y_test)
```

**8. Нормализовать данные и написать модель Random Forest уже на нормализованных данных. Сделать classification report. Сравнить с результатом, полученном в задаче 7. Нужно ли нормализовывать данные?**

```python
headers = list(dataset.columns.values)
```

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

np_scaled = min_max_scaler.fit_transform(dataset)
dataset_norm = pd.DataFrame(np_scaled)
dataset_norm.head()
```

```python
dataset_norm.columns = headers
dataset_norm.head()
```

```python
output_y = dataset_norm.iloc[:,29:]
input_x = dataset_norm.iloc[:,1:29]
```

```python
le=LabelEncoder()
targets=(output_y.iloc[:,:]==1).idxmax(1)
print(targets.value_counts())
Y=le.fit_transform(targets)
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_x, Y, test_size=0.3)
```

```python
# создаем модель деревья решений
# выбираем 100 деревьев в качестве параметра
model=RandomForestClassifier(n_estimators=100)

# обучаем модель
model.fit(X_train, y_train)
```

```python
headers = list(X_train.columns.values)

feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(6, 6))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Важность атрибутов')
plt.ylabel('Атрибуты')
plt.title("Наиболее важные атрибуты")
plt.legend()
plt.show()
```

```python
model_pred = model.predict(X_test)

print(classification_report(y_test, model_pred))

model.score(X_test, y_test)
```

Нормализация немного улучшает качество модели, хотя в данном случае и без нормализации качество было неплохое, а модель Random Forest не чувствительна к масштабированию.

Общая точность модели:

После нормализации 0.9994934919606573
