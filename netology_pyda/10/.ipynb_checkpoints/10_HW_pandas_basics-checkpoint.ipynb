## Задание 1 ##
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера.
Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
%%time

ratings = pd.read_csv('C:/ml-25m/ratings.csv')
```

```python
ratings.head()
```

```python
five_ratings = ratings.query('rating == {}'.format(5.0))
```

```python
five_ratings.head()
```

```python
best_movie = five_ratings.groupby('movieId').count().sort_values('rating', ascending=False).reset_index().head(1)
```

```python
%%time

movies = pd.read_csv('C:/ml-25m/movies.csv')
joined = best_movie.merge(movies, on='movieId', how='left')
```

## Задание 2 ##
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
electricity_consumption = pd.read_csv('power.csv')
electricity_consumption.head()
```

```python
electricity_consumption[ electricity_consumption['country'].str.contains('ia', case=False) ]['country'].unique()
```

```python
#выделяем страны Балтии

baltic = electricity_consumption[
        (electricity_consumption.country == 'Estonia') |
        (electricity_consumption.country == 'Latvia') |
        (electricity_consumption.country == 'Lithuania')
     ]
```

```python
baltic.info()
```

```python
#фильтруем категории

baltic = baltic[ (baltic.category == 4) |
                 (baltic.category == 12) |
                (baltic.category == 21)
     ]
```

```python
#фильтруем года

baltic = baltic[baltic.year >= 2005]
baltic = baltic[baltic.year <= 2010]
```

```python
#оставляем только положительные значения

baltic = baltic[baltic.quantity >= 0]
```

```python
#считаем совокупное энергопотребление

print('Совокупное энергопотребление стран Прибалтики с 2005 по 2010 гг.: ', baltic.quantity.sum(), sep='')
```
