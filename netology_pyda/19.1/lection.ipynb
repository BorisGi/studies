```python
new_size_df = woman_shoes['prices.size'].str.extract(r'([0-9\.]+)? ?([A-Za-z]+)?')
new_size_df.columns = ['size', 'size_cat']
new_size_df.head(5)

woman_shoes = woman_shoes.merge(new_size_df, left_index=True, right_index=True)
woman_shoes.head(5)
```

```python
nike_size_pivot = woman_shoes.query('brand == "nike"').pivot_table(index='name',
                                                                   columns='size',
                                                                   values='avg_price',
                                                                   aggfunc='mean')
nike_size_pivot
```

```python
nike_size_pivot.median(axis=1)
```

```python
m = nike_size_pivot.median(axis=1)
for i, col in enumerate(nike_size_pivot):
    nike_size_pivot.iloc[:, i] = nike_size_pivot.iloc[:, i].fillna(m)
```
